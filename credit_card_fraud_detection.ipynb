```python
import sklearn
import numpy
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\acer\Downloads\credit_card_fraud_detection\creditcard.csv\creditcard.csv')
df.head(20)
```

```python
df = df.drop(columns=['Time'])
df.head(10)
```

```python
df.isnull().sum()
```

```python
df.duplicated().any()
df=df.drop_duplicates()
```

```python
df['Class'] .value_counts()
```

```python

plt.figure(figsize=(12, 8))
plt.hist(df[df['Class'] == 0]['Class'], bins=2,edgecolor='black', alpha=1, label='Normal')
plt.hist(df[df['Class'] == 1]['Class'], bins=2,edgecolor='black', alpha=1, label='Fraud')
plt.xticks([0, 1], ['Normal', 'Fraud'])
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
s=StandardScaler()
df['Amount'] = s.fit_transform(pd.DataFrame(df['Amount']))
df.head(5)
```

```python
numeric_df = df.select_dtypes(include=[np.number])

corr_m = numeric_df.corr()
print(corr_m)
```

```python
corr_m = numeric_df.corr()
fig, ax = plt.subplots(figsize=(40,40))
sns.heatmap(corr_m , annot=True, ax=ax , cmap= 'coolwarm')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
import pickle

X= df.drop(columns=['Class'])
Y= df['Class']
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.30)
```

```python
mode = LogisticRegression()
mode.fit(x_train,y_train)
y_pred = mode.predict(x_test)
print("LogisticRegression accracy: ", mode.score(x_test,y_test)*100)
print("LogisticRegression Classification Report:\n", classification_report(y_test, y_pred))
```

```python
dt_model = DecisionTreeClassifier()
dt_model.fit(x_train,y_train)

y_pred_dt = dt_model.predict(x_test)

print("Decision Tree Accuracy:", accuracy_score(y_test, y_pred_dt))
print("Decision Tree Classification Report:\n", classification_report(y_test, y_pred_dt))
```

```python
import pickle
filename= 'model.sav'
pickle.dump(dt_model, open(filename, 'wb'))
```
